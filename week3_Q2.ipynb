```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pylab as py
```

```python
df=pd.read_csv('TB_burden_countries_2014-09-29.csv')
N=(df.e_inc_tbhiv_num-df.e_inc_tbhiv_num.mean())/(df.e_inc_tbhiv_num.std())
sm.qqplot(N,line='45')
```

```python
plt.hist(N)
plt.show()
```
