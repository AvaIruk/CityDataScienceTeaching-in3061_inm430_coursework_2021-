```python
import numpy as np
from random import sample
from scipy.stats import kurtosis
from scipy.stats import skew
```

```python
S=[2,4,6]
for i in S:
    print('variance = ')
    print(i)
    R=np.random.normal(10,i,size=(1000))
    X=np.arange(0,1000,1)
    S5=sample(list(X),5)
    S10=sample(list(X),10)
    S100=sample(list(X),100)
    M5=np.mean(R[S5])
    print("Mean for 5 samples : ")
    print(M5)
    M10=np.mean(R[S10])
    print("Mean for 10 samples : ")
    print(M10)
    M100=np.mean(R[S100])
    print("Mean for 100 samples : ")
    print(M100)
    print('-----------------------------------')
    ST5=np.std(R[S5])
    print("standard deviation for 5 samples : ")
    print(ST5)
    ST10=np.std(R[S10])
    print("standard deviation for 10 samples : ")
    print(ST10)
    ST100=np.std(R[S100])
    print("standard deviation for 100 samples : ")
    print(ST100)
    print('-----------------------------------')
    K5=kurtosis(R[S5])
    print("kurtosis for 5 samples : ")
    print(K5)
    K10=kurtosis(R[S10])
    print("kurtosis for 10 samples : ")
    print(K10)
    K100=kurtosis(R[S100])
    print("kurtosis for 100 samples : ")
    print(K100)
    print('-----------------------------------')
    SK5=skew(R[S5])
    print("skewness for 5 samples : ")
    print(SK5)
    SK10=skew(R[S10])
    print("skewness for 10 samples : ")
    print(SK10)
    SK100=skew(R[S100])
    print("skewness for 100 samples : ")
    print(SK100)
    print('*************************************')
```
